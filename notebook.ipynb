```python
'''  To Dos   '''
# create a search function to search for specific team
# to search for specific time
# to seach specific competition
# Change time to BST

import bs4
import requests
from datetime import datetime, timedelta
import pprint
from urllib.parse import urljoin

pp = pprint.PrettyPrinter()
```

```python
url = 'https://www.goal.com/en-in/all-competitions'
BASE_URL = 'https://www.goal.com'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246'
}
```

```python
res = requests.get(url, headers=headers)
soup = bs4.BeautifulSoup(res.text, 'lxml')
```

```python
soup
```

```python
x = soup.find('li', class_='widget-navigation__item widget-navigation__top-item')
print(x)
```

```python
def searchAllCompetition(tag):
    return tag.has_attr('href') and tag.get('href') == '/en-in/all-competitions'
```

```python
x = soup.find(searchAllCompetition)
print(x)
```

```python
dir(x)
```

```python
leagueContainerSoup = x.findNextSibling()
```

```python
leagueSoupList = leagueContainerSoup.find('ul').find_all('li')
```

```python
for leagueSoup in leagueSoupList:
    title = leagueSoup.find('a').get('title')
    leagueUrl = leagueSoup.find('a').get('href')
    leagueUrl = urljoin(BASE_URL, leagueUrl)
    print(title, leagueUrl)
```
